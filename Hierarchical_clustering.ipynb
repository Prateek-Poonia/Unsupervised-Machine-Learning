Agglomeratrive Clusterring


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
X,y = make_blobs(n_samples = 10,n_features=2,centers = 3,random_state = 42)
print("Shape of the dataset is : ",X.shape)
```

```python
model = AgglomerativeClustering(n_clusters=3,metric = 'euclidean',linkage = 'single')
labels = model.fit_predict(X)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(X[:,0],X[:,1],c=labels, cmap = 'rainbow', edgecolor = 'k')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Agglomerative Clustering')
plt.show()
```

```python
Z = linkage(X,method = 'single')
plt.figure(figsize=(8,5))
dendrogram(Z)
plt.title("Dendrogram for Hierarchical Clustering")
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```
